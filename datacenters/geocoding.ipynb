```python
import pandas as pd

df = pd.read_csv('data_centers.csv', sep=',')
df.drop(columns=['Unnamed: 0'], inplace=True)

df
```

```python
df['country'] = df['country'].str.strip().str.lower()
df['country'] = df['country'].replace('u.s.', 'usa')
df['country'] = df['country'].replace('united kingdom', 'uk')
df.dropna(subset=['country'], inplace=True)
pattern = r'^[a-zA-Z\s]+$'
df_cleaned = df[df['country'].str.contains(pattern, na=False)]
df_cleaned['country'] = df_cleaned['country'].str.replace(r'\bU\.S\.?\b', 'USA', regex=True)
df_cleaned = df_cleaned[~df_cleaned['country'].str.contains(r'[0-9一-龯]', na=False)]
df_cleaned
```

```python
df_cleaned['country'] = df_cleaned['country'].replace('the netherlands', 'netherlands')
df_cleaned['country'] = df_cleaned['country'].replace('united states', 'usa')
df_cleaned['country'] = df_cleaned['country'].replace('deutschland', 'germany')
df_cleaned['country'] = df_cleaned['country'].replace('italia', 'italy')
df_cleaned['country'] = df_cleaned['country'].replace('latvija', 'latvia')
df_cleaned['country'] = df_cleaned['country'].replace('republic of ireland', 'ireland')
df = df_cleaned[df_cleaned['country'] != 'gasan digital center']
df = df[df['country'] != 'apple store']
```

```python
import requests

def geocode(location, index, df) -> None:
    url = f'https://geocode.maps.co/search?q={location}&api_key=668d02c6d9de4757313352qgufc7ba3'
    response = requests.get(url)
    if response.status_code == 200 and response.json():
        print(response.json()[0])
        df.loc[index, 'latitude'] = float(response.json()[0]['lat'])
        df.loc[index, 'longitude'] = float(response.json()[0]['lon'])
```

```python
import numpy as np

df['latitude'] = np.nan
df['longitude'] = np.nan
df
```

```python
import time

for index in range(len(df)):
    try:
        location: str = f'{df.loc[index, "city"]}, {df.loc[index, "country"]}'
    except:
        continue
    geocode(location, index)
    time.sleep(1.5)

df
```

```python
df.to_csv('geo_data_centers.csv')
```
